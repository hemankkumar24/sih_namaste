```python
ayurveda_df = ayurveda_df[ayurveda_df['Long_definition'].notna()].copy()
siddha_df = siddha_df[siddha_df['Long_definition'].notna()].copy()
unani_df = unani_df[unani_df['Long_definition'].notna()].copy()
```

```python
ayurveda_df.shape, siddha_df.shape, unani_df.shape
```

```python
icd11_df = pd.read_excel("icd11.xlsx")
```

```python
icd11_df.head()
```

```python
icd11_df['Title'] = icd11_df['Title'].str.replace("-", "", regex=False).str.strip()
```

```python
icd11_df.head()
```

```python
from sentence_transformers import SentenceTransformer, util
```

```python
model = SentenceTransformer("all-MiniLM-L6-v2")
```

```python
ayurveda_emb = model.encode(ayurveda_df["Long_definition"].tolist(), convert_to_tensor=True)
siddha_emb   = model.encode(siddha_df["Long_definition"].tolist(), convert_to_tensor=True)
unani_emb    = model.encode(unani_df["Long_definition"].tolist(), convert_to_tensor=True)
icd11_emb    = model.encode(icd11_df["Title"].tolist(), convert_to_tensor=True)
```

```python
sim_icd11_ayur   = util.cos_sim(icd11_emb, ayurveda_emb)
sim_icd11_siddha = util.cos_sim(icd11_emb, siddha_emb)
sim_icd11_unani  = util.cos_sim(icd11_emb, unani_emb)
```

```python
ayur_best_idx   = sim_icd11_ayur.argmax(dim=1).cpu().numpy()
siddha_best_idx = sim_icd11_siddha.argmax(dim=1).cpu().numpy()
unani_best_idx  = sim_icd11_unani.argmax(dim=1).cpu().numpy()
```

```python
combined_rows = []
for i, row in icd11_df.iterrows():
    ayur_match   = ayurveda_df.iloc[ayur_best_idx[i]].to_dict()
    siddha_match = siddha_df.iloc[siddha_best_idx[i]].to_dict()
    unani_match  = unani_df.iloc[unani_best_idx[i]].to_dict()
    icd11_match = row.to_dict()
    
    combined_rows.append({
        **{f"ICD11_{k}": v for k, v in icd11_match.items()},

        **{f"Ayurveda_{k}": v for k, v in ayur_match.items()},
        **{f"Siddha_{k}": v for k, v in siddha_match.items()},
        **{f"Unani_{k}": v for k, v in unani_match.items()},

        "Ayurveda_Similarity": float(sim_icd11_ayur[i, ayur_best_idx[i]]),
        "Siddha_Similarity": float(sim_icd11_siddha[i, siddha_best_idx[i]]),
        "Unani_Similarity": float(sim_icd11_unani[i, unani_best_idx[i]])
    })

combined_df = pd.DataFrame(combined_rows)
combined_df.to_csv("icd11_as_anchor.csv", index=False)
```
